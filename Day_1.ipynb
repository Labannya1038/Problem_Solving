# Calulation sign and usees:
"+" Add two or multiple values

"-" Substract two value

"*" Multiplication two or multiple values

"/" Division two values

"%" Modulo (5%2 = 1); Which gives reminder values

"**" Exponential or Power

">" Greater Than (6>5)

">=" Greater or equal

"<" Less Than (5<6)

"<=" Less or equal

"==" Equal to (5==5)

"!=" Not equal to (5 != 6)


```python
print(4+8)
print(8-4)
print(4*8)
print(8/4)
print(4%8)
print(9**4)
print(4>8)
print(4>=8)
print(4<8)
print(4<=8)
print(4==8)
print(4==4)
print(4!=8)
print(4!=4)
```
