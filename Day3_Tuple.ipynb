# **Tuple**

- Tuples are an ordered sequence.

- Tuples are written as comma separated elements within parenthesis.

- Tuples are immutable in nature i.e. the elements in the tuple cannot be added or removed once created.

```python
tuple_1= (1,2,3,4)
```

```python
type(tuple_1)
```

```python
tuple_1[0]=5  #immutable
```

```python
tuple_1.append(5)
```

```python
del(tuple_1[0])
```

- We can combined the Tuple by adding.

```python
A =(1,2,3)
B =(7,8,9)
C = A+B
```

# **tuple method**
- count() Returns the number of times a specified value occurs in a tuple
- index() Searches the tuple for a specified value and returns the position of where it was found
- count() Returns the number of times a specified value occurs in a tuple
- syntax: tuple.count(value)

```python
data_3 = (10, 20 ,30, 10, 40, 20, 10)
```

```python
data_3.count(20)
```

```python
data_3.count(10)
```

```python
data_3.index(10)
```
